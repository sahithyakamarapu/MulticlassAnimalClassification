```python
import os
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid")
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report
```

```python
# Path to the dataset
dataset_path = "/root/.cache/kagglehub/datasets/iamsouravbanerjee/animal-image-dataset-90-different-animals/versions/5/animals/animals"

# List all the classes (Animal Types)
classes = os.listdir(dataset_path)
num_classes = len(classes)

# Display the number of classes along with names
print(f'Number of Classes: {num_classes}')
print(f'Classes: {classes}')
```

```python
# Visualize some sample images form the dataset
plt.figure(figsize=(15,10))
for i, class_name in enumerate(classes[:10]):
  class_path = os.path.join(dataset_path, class_name)
  img_name = os.listdir(class_path)[0]
  img_path = os.path.join(class_path, img_name)
  img = plt.imread(img_path)

  plt.subplot(2, 5, i+1)
  plt.imshow(img)
  plt.suptitle("Sample Images of 10 Animals", fontsize=20, fontweight='bold')
  plt.title(f'{class_name} \n shape: {img.shape}')
  plt.axis('off')
plt.show()
```

```python
# Filter class names that start with 'b' or 'B'
b_animals = [cls for cls in classes if cls.lower().startswith('b')]

print(f"Animals that start with 'B': {b_animals}")
print(f"Total: {len(b_animals)}")
```

```python
import matplotlib.pyplot as plt
import os
import random
from PIL import Image

# Filter class names that start with 'b' or 'B'
b_animals = [cls for cls in classes if cls.lower().startswith('b')]

# Visualize
plt.figure(figsize=(15, 5))

for i, class_name in enumerate(b_animals[:5]):  # Limit to 5 for layout
    class_path = os.path.join(dataset_path, class_name)
    img_list = os.listdir(class_path)

    if not img_list:
        continue

    img_name = random.choice(img_list)
    img_path = os.path.join(class_path, img_name)

    try:
        img = Image.open(img_path).resize((128, 128))
        plt.subplot(1, 5, i + 1)
        plt.imshow(img)
        plt.title(f'{class_name}', fontsize=12)
        plt.axis('off')
    except Exception as e:
        print(f"Error loading image from {img_path}: {e}")

plt.suptitle("Sample Images of Animals whose names starting with 'B'", fontsize=18, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.9])
plt.show()
```
